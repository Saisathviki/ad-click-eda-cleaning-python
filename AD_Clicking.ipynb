```python
print("..............FIRST MOMENT BUSINESS DECISION...............")
print("\n")
print('Calculating Mean')
print(f"Mean of Daily_Time_Spent_on_Site is:{ad_data.Daily_Time_Spent_on_Site.mean()}")
print(f"Mean of Age is:{ad_data.Age.mean()}")
print(f"Mean of Area_Income is:{ad_data.Area_Income.mean()}")
print(f"Mean of Daily_Internet_Usage is:{ad_data.Daily_Internet_Usage.mean()}")
print(f"Mean of Male is:{ad_data.Male.mean()}")
print(f"Mean of Clicked_on_Ad is:{ad_data.Clicked_on_Ad.mean()}")
print("\n\n")
print('Calculating Median')
print(f"Median of Daily_Time_Spent_on_Site is:{ad_data.Daily_Time_Spent_on_Site.median()}")
print(f"Median of Age is:{ad_data.Age.median()}")
print(f"Median of Area_Income is:{ad_data.Area_Income.median()}")
print(f"Median of Daily_Internet_Usage is:{ad_data.Daily_Internet_Usage.median()}")
print(f"Median of Male is:{ad_data.Male.median()}")
print(f"Median of Clicked_on_Ad is:{ad_data.Clicked_on_Ad.median()}")
print("\n\n")
print('Calculating Mode')
print(f"Mode of Daily_Time_Spent_on_Site is:{ad_data.Daily_Time_Spent_on_Site.mode()}")
print(f"Mode of Age is:{ad_data.Age.mode()}")
print(f"Mode of Area_Income is:{ad_data.Area_Income.mode()}")
print(f"Mode of Daily_Internet_Usage is:{ad_data.Daily_Internet_Usage.mode()}")
print(f"Mode of Ad_Topic_Line is:{ad_data.Ad_Topic_Line.mode()}")
print(f"Mode of City is:{ad_data.City.mode()}")
print(f"Mode of Male is:{ad_data.Male.mode()}")
print(f"Mode of Clicked_on_Ad is:{ad_data.Clicked_on_Ad.mode()}")
print("\n\n")
```

```python
# Second Moment Business Decision (Calculates Spread)
print("...............SECOND MOMENT BUSINESS DECISION.............. ")
print("Calculating Variance")
print(f"Variance of Daily_Time_Spent_on_Site is:{ad_data.Daily_Time_Spent_on_Site.var()}")
print(f"Variance of Age is:{ad_data.Age.var()}")
print(f"Variance of Area_Income is:{ad_data.Area_Income.var()}")
print(f"Variance of Daily_Internet_Usage is:{ad_data.Daily_Internet_Usage.var()}")
print(f"Variance of Male is:{ad_data.Male.var()}")
print(f"Variance of Clicked_on_Ad is:{ad_data.Clicked_on_Ad.var()}")
print("\n\n")
print("Calculating Standard Deviation")
print(f"Standard deviation of Daily_Time_Spent_on_Site is:{ad_data.Daily_Time_Spent_on_Site.std()}")
print(f"Standard deviation of Age is:{ad_data.Age.std()}")
print(f"Standard deviation of Area_Income is:{ad_data.Area_Income.std()}")
print(f"Standard deviation of Daily_Internet_Usage is:{ad_data.Daily_Internet_Usage.std()}")
print(f"Standard deviation of Male is:{ad_data.Male.std()}")
print(f"Standard deviation of Clicked_on_Ad is:{ad_data.Clicked_on_Ad.std()}")
print("\n\n")
print("Calculating range")
print(f"range of Daily_Time_Spent_on_Site is:{max(ad_data.Daily_Time_Spent_on_Site)-min(ad_data.Daily_Time_Spent_on_Site)}")
print(f"Range of Age is:{max(ad_data.Age)-min(ad_data.Age)}")
print(f"Range of Area_Income is:{max(ad_data.Area_Income)-min(ad_data.Area_Income)}")
print(f"Range of Daily_Internet_Usage is:{max(ad_data.Daily_Internet_Usage)-min(ad_data.Daily_Internet_Usage)}")
print(f"Range of Male is:{max(ad_data.Male)-min(ad_data.Male)}")
print(f"Range of Clicked_on_Ad is:{max(ad_data.Clicked_on_Ad)-min(ad_data.Clicked_on_Ad)}")
print("\n\n")

# Max & Min values are not a part of second decision
print("Maximum and Minimum Values:")
print(max(ad_data.Daily_Time_Spent_on_Site))
print(min(ad_data.Daily_Time_Spent_on_Site))
print(max(ad_data.Age))
print(min(ad_data.Age))
print(max(ad_data.Area_Income))
print(min(ad_data.Area_Income))
print(max(ad_data.Daily_Internet_Usage))
print(min(ad_data.Daily_Internet_Usage))
print(max(ad_data.Male))
print(min(ad_data.Male))
print(max(ad_data.Clicked_on_Ad))
print(min(ad_data.Clicked_on_Ad))
```

```python
# Third Moment Business Decision (Calculates Skewness(Asymmetric Distribution)) - (Mean<Median<Mode)coz mean is affected by outliers.
print(".............THIRD MOMENT BUSINESS DECISION....................")
print("Calculate Skewness")
print(f"Skewness of Daily_Time_Spent_on_Site is:{ad_data.Daily_Time_Spent_on_Site.skew()}")
print(f"skewness of Age is:{ad_data.Age.skew()}")
print(f"skewness of Area_Income is:{ad_data.Area_Income.skew()}")
print(f"skewness of Daily_Internet_Usage is:{ad_data.Daily_Internet_Usage.skew()}")
print(f"skewness of Male is:{ad_data.Male.skew()}")
print(f"skewness of Clicked_on_Ad is:{ad_data.Clicked_on_Ad.skew()}")
print("\n\n")
```

```python
# Fourth Moment Business Decision (Kurtosis/Peakedness)
# sharp peak - (+)ve kurtosis - Leptokurtic
# wide peak - (-)ve kurtosis - Platykurtic

print("................FOURTH MOMENT BUSINESS DECISION......................")
print("kurtosis")
print(f"Kurtosis of Daily_Time_Spent_on_Site is:{ad_data.Daily_Time_Spent_on_Site.kurt()}")
print(f"Kurtosis of Age is:{ad_data.Age.kurt()}")
print(f"Kurtosis of Area_Income is:{ad_data.Area_Income.kurt()}")
print(f"Kurtosis of Daily_Internet_Usage is:{ad_data.Daily_Internet_Usage.kurt()}")
print(f"Kurtosis of Male is:{ad_data.Male.kurt()}")
print(f"Kurtosis of Clicked_on_Ad is:{ad_data.Clicked_on_Ad.kurt()}")
```

# Handling missing values

```python

import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# Check DataFrame structure and missing values
ad_data.info()
print(ad_data.isna().sum())

# Impute missing values with mean for 'Daily_Time_Spent_on_Site'
mean_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
ad_data["Daily_Time_Spent_on_Site"] = mean_imputer.fit_transform(ad_data[["Daily_Time_Spent_on_Site"]]).ravel()
print("Missing values in 'Daily_Time_Spent_on_Site':", ad_data["Daily_Time_Spent_on_Site"].isna().sum())

# Impute missing values with median for 'Age'
median_imputer = SimpleImputer(missing_values=np.nan, strategy='median')
ad_data["Age"] = median_imputer.fit_transform(ad_data[["Age"]]).ravel()
print("Missing values in 'Age':", ad_data["Age"].isna().sum())

# Impute missing values with most frequent (mode) for 'Male'
mode_imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
ad_data["Male"] = mode_imputer.fit_transform(ad_data[["Male"]]).ravel()
print("Missing values in 'Male':", ad_data["Male"].isna().sum())
```
